```python
import requests
import pandas as pd
import re
```

```python
url = "https://www.marvel.com/movies"
content = requests.get(url).text
pattern = "card-body__headline\">(.+?)</p>.*?(\d{4})</p>.*?"
matched = re.findall(pattern, content, re.S)
for m in matched:
    print(f"{m[1]}")
```

```python
pd.DataFrame(matched, columns=["movie", "year"])
```
